```python
import time
import builtwith
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import WebDriverException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
```

```python
builtwith.parse('https://www.greenfinder-mobility.com/e-bikes/models/')
```

```python
def init_driver():
    driver=webdriver.Chrome()
    driver.wait = WebDriverWait(driver, 10)
    return driver
```

```python
def lookup(driver, timeout):
    driver.get("https://www.greenfinder-mobility.com/e-bikes/models/")
    scroll_pause_time = timeout
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        try:
            element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, "loadmore")))
        except TimeoutException:
            soup=BS(driver.page_source, 'lxml')
            return soup
        try:
            time.sleep(scroll_pause_time)
            element.click()
        except WebDriverException:
            soup=BS(driver.page_source, 'lxml')
            return soup
        except KeyboardInterrupt:
            soup=BS(driver.page_source, 'lxml')
            return soup
```

```python
driver = init_driver()
soup=lookup(driver, timeout=5)
time.sleep(5)
driver.quit()
#there are a lot of pages, so I stopped the cycle
```

```python
def get_year(soup):
    table=soup.find('table', width="100%")
    tr=table.findAll('tr')[0]
    return tr.findAll('td')[1].text
```

```python
model=[]
year=[]
price=[]
category=[]
bike_type=[]
frame=[]
size=[]
weight=[]
max_weight=[]
bike_range=[]
```

```python
for i in links[0:30]:
    time.sleep(3)
    html=requests.get(i, headers=headers)
    soup=BS(html.text, 'lxml')
    model.append(get_model(soup))
    year.append(get_year(soup))
    price.append(get_price(soup))
    category.append(get_category(soup))
    bike_type.append(get_type(soup))
    frame.append(get_frame(soup))
    size.append(get_size(soup))
    weight.append(get_weight(soup))
    max_weight.append(get_max_weight(soup))
    bike_range.append(get_range(soup))
```

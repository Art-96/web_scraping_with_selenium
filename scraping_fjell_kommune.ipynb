```python
import re
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import WebDriverException
```

```python
def init_driver():
    driver=webdriver.Chrome()
    driver.wait = WebDriverWait(driver, 5)
    return driver

def lookup(driver, timeout):
    driver.get("https://www.bergen.kommune.no/sok?q=barnehager&innholdstype=barnehage")
    scroll_pause_time = timeout
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        button = driver.find_element_by_class_name("read_more_link--20")
        try:
            button.click()
            time.sleep(scroll_pause_time)
        except WebDriverException:
            soup=BeautifulSoup(driver.page_source, 'lxml')
            return soup
```

```python
driver = init_driver()
soup=lookup(driver, timeout=3)
time.sleep(5)
driver.quit()
```

```python
telefon=[tel.text for tel in soup.findAll('div', class_='search_result_excerpt')]
```

```python
telefon
```

```python
number_telefon=[re.findall(r'\s\b[\d\s\/]+\b', str(i)) for i in telefon]
```

```python
number_telefon
```

```python
teli=[]
for i in number_telefon:
    for l in i:
        teli.append(l)

teli[110]+='/'+teli[111]
del(teli[111])
teli.insert(0, 'None')
```

```python
len(teli)
```

```python
li_links=[line for line in soup.findAll('li', class_='barnehage')]
```

```python
li_links
```

```python
line=[]
line2=[]
for i in li_links:
    if i.find('a', class_='null'):
        for l in re.findall(r'[A-Za-z.]+\@', i.text):
            line.append(l)
        for m in re.findall(r'[@a-z.]+\.no|\.com|\.as|\.org', i.text):
            line2.append(m)
    else:
        line.append(None)
        line2.append(None)
```

```python
line
```

```python
line2
```

```python
lmite=[re.sub(r"[@]", "", str(i)) for i in line]
```

```python
res_list = [] 
for i in range(0, len(lmite)):
    if 'None'in str(lmite[i]):
        res_list.append('None')
    else:
        res_list.append(str(lmite[i])+str(line2[i])) 

res_list
```

```python
bd=pd.DataFrame({'Name':name,'Email':res_list,'Telefon':teli,'Links':name_links})
```

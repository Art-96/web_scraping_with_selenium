```python
import time
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import WebDriverException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
```

```python
path = 'C:\\Users\\Admin\\Documents\\chromedriver_win32\\chromedriver.exe'
# path = 'write the path to your webdriver'
```

```python
# initializing the webdriver
def init_driver():
    driver=webdriver.Chrome('C:\\Users\\Admin\\Documents\\chromedriver_win32\\chromedriver.exe')
    driver.wait = WebDriverWait(driver, 5)
    return driver
```

```python
Product_Info=[]
```

```python
def get_product_urls(driver,url):
    driver.get(url)
    time.sleep(5)
    
    element = driver.find_element_by_xpath("//div[@class='__list_view']")
    products_list = element.find_elements_by_xpath("//div[@class='__lot_image']/a")
    

    for i in range(0,len(products_list)):
        elem = driver.find_element_by_xpath("//div[@class='__list_view']")
        prod_list =  elem.find_elements_by_xpath("//div[@class='__lot_image']/a")
        
        prod_list[i].click()
        time.sleep(5)
        
        try:
            Property_Adress1 = driver.find_element_by_xpath("//span[@class='__address']").text
        except:
            Property_Adress1 = None
        
        try:
            Lost_Number = driver.find_element_by_xpath("//h2[@class='__bold __m_0']/span[1]").text
        except:
            Lost_Number = None
        
        try:
            Location = driver.find_element_by_xpath("//h2[@class='__bold __m_0']/span[2]").text.strip('-')
        except:
            Location = None
        
        try:
            asd = driver.find_element_by_xpath("//a[@class='__block __m_t_5']/span").text.split('-')
            Auction_Type = asd[0]
            Auction_Date = asd[1]
        except:
            Auction_Type = None
            Auction_Date = None
            
        try:
            Type = driver.find_element_by_xpath("//h1[@class='__bold __m_t_0']/span").text
        except:
            Type = None
            
        try:
            Price = driver.find_element_by_xpath("//h3[@class='__m_0 __bold __lot_price pull-left']/span").text
        except:
            Price = None
            
        try:
            Property_Tenure = driver.find_element_by_xpath("//*[@id='tab']/div/div/div/div[1]/div/div[2]/div/div/div[1]/div/p/span/span[1]").text
        except:
            Property_Tenure = None
        
        try:
            Property_Location = ''
            elements = driver.find_elements_by_xpath("//ul[@class='__ul_location']/li/span")
            for l in elements:
                Property_Location += l.text + os.linesep
        except:
            Property_Location = None
        
        try:
            Property_Additional_Information = driver.find_element_by_xpath("//h5[text()='Additional Information']/..//li").text
        except:
            Property_Additional_Information = None
        
        try:
            Property_Tenure = driver.find_element_by_css_selector("div.col-lg-8 div p span span").text
        except:
            Property_Tenure = None
        
        try:
            By_Order_Of = driver.find_element_by_css_selector("div.__vendor_text span").text
        except:
            By_Order_Of = None
        
        try:
            Key_Features = ''
            elems = driver.find_elements_by_css_selector("div.__property_details div ul.__m_b_30 li")
            for el in elems:
                Key_Features += el.text + os.linesep
        except:
            Key_Features = None
            
        try:
            elm2 = driver.find_element_by_xpath("//span[@class='__image_placeholder __image_ratio']/div[@class='__image_div']").get_attribute("style")
            Property_Image = "https://auctions.allsop.co.uk/" + str(elm2.split('"')[1])
        except:
            Property_Image = None
        
        try:
            lot_status_class = driver.find_element_by_xpath("//div[contains(@class, '__property_slider_image')]").get_attribute('class')
            Lot_Status = lot_status_class.split()[-1].replace('_',' ').strip()
        except:
            Lot_Status = None
            

        Property_URL = driver.current_url
        
        driver.back()
        print(pd.DataFrame({'Property_Adress1' : [Property_Adress1], 
                                          'Lost_Number' : [Lost_Number], 
                                          'Location' : [Location],
                                          'Auction_Type':[Auction_Type],
                                          'Auction_Date':[Auction_Date],
                                          'Type':[Type],
                                          'Price':[ Price],
                                          'Lot_Status':[Lot_Status],
                                          'Property_Tenure':[Property_Tenure],
                                          'Property_Location':[Property_Location],
                                          'Property_Additional_Information':[Property_Additional_Information],
                                          'Property_Tenure':[Property_Tenure],
                                          'By_Order_Of':[By_Order_Of],
                                          'Key_Features':[Key_Features],
                                          'Property_Image':[Property_Image],
                                          'Property_URL' : [Property_URL]}))
        Product_Info.append(pd.DataFrame({'Property_Adress1' : [Property_Adress1],
                                          'Lost_Number' : [Lost_Number],
                                          'Location' : [Location],
                                          'Auction_Type':[Auction_Type],
                                          'Auction_Date':[Auction_Date],
                                          'Type':[Type],
                                          'Price':[ Price],
                                          'Lot_Status':[Lot_Status],
                                          'Property_Location':[Property_Location],
                                          'Property_Additional_Information':[Property_Additional_Information],
                                          'Property_Tenure':[Property_Tenure],
                                          'By_Order_Of':[By_Order_Of],
                                          'Key_Features':[Key_Features],
                                          'Property_Image':[Property_Image],
                                          'Property_URL' : [Property_URL]}))
```

```python
driver = init_driver()
elements=get_product_urls(driver, url)
driver.close()
```

```python
final_df = pd.concat(Product_Info)
```

```python
final_df.dropna(subset=['Property_Adress1']).reset_index(drop=True)
```
